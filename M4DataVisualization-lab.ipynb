# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
plt.figure(figsize=(10, 6))
QUERY = """
SELECT ConvertedComp
FROM master
"""
converted_comp = pd.read_sql_query(QUERY, conn)
sns.histplot(converted_comp['ConvertedComp'], bins=30, kde=True)
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
plt.figure(figsize=(8, 6))
QUERY = """
SELECT Age
FROM master
"""
age_data = pd.read_sql_query(QUERY, conn)
sns.boxplot(x='Age', data=age_data)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.figure(figsize=(8, 6))
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
age_workweek = pd.read_sql_query(QUERY, conn)
sns.scatterplot(x='Age', y='WorkWeekHrs', data=age_workweek)
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.figure(figsize=(10, 8))
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
bubble_data = pd.read_sql_query(QUERY, conn)
sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', size='Age', sizes=(20, 200), data=bubble_data, hue ="Age")
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs (Bubble size: Age)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
plt.figure(figsize=(8, 8))
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(Respondent) AS Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
top_databases = pd.read_sql_query(QUERY, conn)
plt.pie(top_databases['Count'], labels=top_databases['DatabaseDesireNextYear'], autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

plt.figure(figsize=(8, 6))
QUERY = """
SELECT Age, 
    MAX(CASE WHEN Quartile = 2 THEN WorkWeekHrs END) AS Median_WorkWeekHrs,
    MAX(CASE WHEN Quartile1 = 2 THEN CodeRevHrs END) AS Median_CodeRevHrs
FROM (
    SELECT Age, WorkWeekHrs, CodeRevHrs,
           NTILE(4) OVER(ORDER BY WorkWeekHrs) AS Quartile,
           NTILE(4) OVER(ORDER BY CodeRevHrs) AS Quartile1
    FROM master
    WHERE Age BETWEEN 30 AND 35
)
GROUP BY Age
"""
stacked_data = pd.read_sql_query(QUERY, conn)
stacked_data.plot(kind='bar', x='Age', stacked=True, figsize=(10, 6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
plt.figure(figsize=(8, 6))
QUERY = """
SELECT Age, MAX(CASE WHEN Quartile1 = 2 THEN ConvertedComp END) AS Median_ConvertedComp
FROM (
    SELECT Age, ConvertedComp,
           NTILE(4) OVER(ORDER BY ConvertedComp) AS Quartile1
    FROM master
    WHERE Age BETWEEN 45 AND 60
)
GROUP BY Age
"""
line_data = pd.read_sql_query(QUERY, conn)
plt.plot(line_data['Age'], line_data['Median_ConvertedComp'], marker='o')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
plt.figure(figsize=(10, 6))
QUERY = """
SELECT MainBranch, COUNT(*) AS Count
FROM master
GROUP BY MainBranch
"""
main_branch_data = pd.read_sql_query(QUERY, conn)
main_branch_data.set_index('MainBranch')['Count'].plot(kind='barh')
plt.title('Main Branch Distribution')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.show()
```
